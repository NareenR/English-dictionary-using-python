```python
import os
import json
h="word.txt"
g=os.path.isfile(h)
j=os.path.getsize(h)
def menu():
    menu = {}
    menu['1']="Add new word and its meaning" 
    menu['2']="Find the meaning"
    menu['3']="Update a word"
    menu['4']="Exit"
    while True: 
        options=menu.keys()
        for entry in options: 
            print(entry, menu[entry])

        selection=input("Please Select:") 
        if selection =='1': 
            add_word_and_its_meaning()
        elif selection == '2':
            find_meaning()
        elif selection == '3':
            update_a_word()
        elif selection == '4':
            print("<<<<---->>>>")
            break
        else: 
            print("Unknown Option Selected!") 
def add_word_and_its_meaning():
    if g==False or j==0:
        adding_word()
    else:
        adding_word_to_existing_file()
def find_meaning():
    with open(h,'r') as f:
        words=json.load(f)
        w=input("enter the word:")
        print("Finding",w,"...")
        if w in words:
            meaning=words.get(w)
            print("meaning of",w,"is:",meaning)
        else:
            print(w,"does not exist in the dictionary")
    
def adding_word_to_existing_file():
    f=open(h,'r')
    words=json.load(f)
    f.close()
    n=int(input('how many words to be added: '))
    i=1
    while i<=n:
        word=input('enter a word: ')
        meaning=input('enter a string: ')
        print("Adding...")
        word.lower()
        meaning.lower()
        words.update({word:meaning})
        i+=1
    f=open(h,'w')
    json.dump(words,f)
    f.close()
def adding_word():
    n=int(input("how many words to be added: "))
    i=1
    Dict={}
    while i<=n:
        word=input('enter a word: ')
        meaning=input('enter a string: ')
        print("Adding...")
        word.lower()
        meaning.lower()
        Dict[word] = meaning
        i+=1
    f=open(h,'w')
    json.dump(Dict,f)
    f.close()
def updating():
    f=open(h,'r')
    words=json.load(f)
    f.close()
    word=input("enter a word: ")
    meaning=input('enter a string: ')
    word.lower()
    meaning.lower()
    append_value(words,word,meaning)
    f=open(h,'w')
    json.dump(words,f)
    f.close()
def append_value(dict_obj, key, value):
    if key in dict_obj:
        if not isinstance(dict_obj[key], list):
            dict_obj[key] = [dict_obj[key]]
        dict_obj[key].append(value)
    else:
        dict_obj[key] = value
def delete():
    f=open(h,'r')
    words=json.load(f)
    f.close()
    word=input("enter a word: ")
    del words[word]
    f=open(h,'w')
    json.dump(words,f)
    f.close()
def update_a_word():
    print("-------type 1 to update a word--------\n------type 2 to delete a word-------")
    n=int(input("enter a option: "))
    if n==1:
        updating()
    elif n==2:
        delete()
    else:
        print("invalid option")
menu()
```
